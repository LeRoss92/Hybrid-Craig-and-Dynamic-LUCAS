```python
import pickle

with open("5_selection_meta.pkl", "rb") as f:
    selection_meta_df = pickle.load(f)

with open("5_with_predictions.pkl", "rb") as f:
    out_df = pickle.load(f)
```

```python
for col in out_df.columns.values:
    print(col)
```

```python
from config import TARGET_CONFIG, use_model

for var in use_model:
    for inference in TARGET_CONFIG[var]["inference"]:
        columns = [f"pred_{var}_fold{f}_{use_model[var]}_inf_{inference['target_name'].replace(' ', '_')}" for f in range(10)]
        all_folds = out_df[columns]
        out_df[f"pred_{var}_median_{use_model[var]}_inf_{inference['target_name'].replace(' ', '_')}"] = all_folds.median(axis=1)
        print('new:', f"pred_{var}_median_{use_model[var]}_inf_{inference['target_name'].replace(' ', '_')}")
```

```python
with open("5_with_predictions.pkl", "wb") as f:
    pickle.dump(out_df, f)
```
